```python
import pandas as pd

from padelpy import from_smiles

from rdkit import Chem
from tqdm.notebook import trange, tqdm


df = pd.read_csv('../smiles-transformer-master/data/BBB_dataset_refined.csv')
smiles = list(df['Canonical SMILES'])

descriptors = pd.DataFrame()

for i in tqdm(range(len(smiles))):
    smiles[i] = smiles[i].replace('\n',"")
    smiles[i] = smiles[i].replace(' ',"")
    
    try:
        desc_fp = from_smiles(smiles[i], fingerprints=False)
    except:
        print("error on smiles:",i, smiles[i])
    if i==0:
        descriptors = pd.DataFrame(desc_fp, index=[0])
    else:  
        descriptors.loc[i] = list(desc_fp.values())

```
